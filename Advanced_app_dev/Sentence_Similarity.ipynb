```python
# Example to use the 'transformers' library for text classificationfrom transformers import pipeline
from sentence_transformers import SentenceTransformer
import torch

# Load model
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
# Encode sentences and compute similarity score
# Define two example sentences
sentence1 = "The movie is good."
sentence2 = "The movie is great."
```

```python
# Encode sentences and compute similarity score
embeddings1 = model.encode([sentence1], convert_to_tensor=True)
embeddings2 = model.encode([sentence2], convert_to_tensor=True)
cosine_similarities = torch.nn.functional.cosine_similarity(embeddings1, embeddings2)
```

```python
cosine_similarities
```

```python
def compute_score(sentence1,sentence2):
    embeddings1 = model.encode([sentence1], convert_to_tensor=True)
    embeddings2 = model.encode([sentence2], convert_to_tensor=True)
    cosine_similarities = torch.nn.functional.cosine_similarity(embeddings1, embeddings2)
    return cosine_similarities.item()


```

#Gradio Interface

```python
# Create a Gradio interface to interactively demonstrate the image captioning functionality
import gradio as gr


gr.close_all()
demo = gr.Interface(fn=compute_score,
                    inputs=[gr.Textbox(label="1st Sentence", lines=2),gr.Textbox(label="2nd Sentence", lines=2)],
                    outputs=[gr.Textbox(label="Similarity Score", lines=2)],

                    allow_flagging="never",
                    #Here we introduce a new tag, examples, easy to use examples for your application
)
demo.launch(share=True)
```

```python
# Close all Gradio interfaces after launching the demo
gr.close_all()
```
