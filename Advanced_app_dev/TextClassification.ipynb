```python
import os
import io
import IPython
from IPython.display import Image, display, HTML
from PIL import Image
import base64

```

```python
# Example to use the 'transformers' library for text classification
from transformers import pipeline

# Load the 'sentiment-analysis' pipeline
model = pipeline("sentiment-analysis")

```

```python
# Encode sentences and compute similarity score
# Define two example sentences
sentence1 = "The movie is good."
sentence2 = "The movie is great."
```

```python
model(sentence1)
```

```python
def classify(input):
  outputs = model(input)

  return f'Label: {outputs[0]["label"]}, Score: {outputs[0]["score"]}'
```

#Gradio Interface

```python
# Create a Gradio interface to interactively demonstrate the image captioning functionality
import gradio as gr


gr.close_all()
demo = gr.Interface(fn=classify,
                    inputs=[gr.Textbox(label="Review:", lines=2)],
                    outputs=[gr.Textbox(label="Sentiment", lines=2)],

                    allow_flagging="never",
                    #Here we introduce a new tag, examples, easy to use examples for your application
)
demo.launch(share=True)
```

```python
# Close all Gradio interfaces after launching the demo
gr.close_all()
```
