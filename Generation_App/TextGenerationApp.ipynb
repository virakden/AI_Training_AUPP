```python
#Helper function
from transformers import pipeline

pipe = pipeline('text-generation')

```

```python
pipe('How large is the earth ?')
```

# Generating the app with Gradio

```python
import gradio as gr


def generate(prompt):
    output = pipe(prompt)
    result_text = output[0]['generated_text']
    return result_text

gr.close_all()
```

```python
with gr.Blocks() as demo:
    gr.Markdown("""# Text completion""")
    with gr.Row():
        with gr.Column(scale = 4):
            prompt = gr.Textbox(label = 'Your prompt')

        with gr.Column(scale = 1, min_width = 50):
            btn = gr.Button('Submit')
            output = gr.Text(label = 'Result')


    btn.click(fn = generate, inputs = [prompt], outputs = [output])
gr.close_all()
demo.launch(share = True,debug=True)
```
