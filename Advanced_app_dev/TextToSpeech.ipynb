```python
import os
import io
import IPython
from IPython.display import Image, display, HTML
from PIL import Image
import base64

```

```python
# Example to use the 'transformers' library for text classificationfrom transformers import pipeline
from transformers import pipeline

# Load model
model = pipeline('text-to-speech')
```

```python
# Encode sentences and compute similarity score
# Define two example sentences
sentence1 = "The movie is good."
```

Supported languages

https://huggingface.co/facebook/nllb-200-distilled-600M/blob/main/README.md

```python
results = model(sentence1)
```

```python
results ["audio"].shape
```

```python
from IPython.display import Audio
Audio(results ["audio"], rate=results ["sampling_rate"])
```

```python
import tempfile
import uuid
import os
from scipy.io.wavfile import write
def translate(input):
    results = model(input)
    file_id = uuid.uuid1()
    file_path = os.path.join(
        tempfile.gettempdir(),
        f'{file_id}.wav'
    )
    write(file_path, rate=results ["sampling_rate"], data=results ["audio"].squeeze())
    return file_path
```

```python
translate('the movie is good.')
```

#Gradio Interface

```python
# Create a Gradio interface to interactively demonstrate the image captioning functionality
import gradio as gr


gr.close_all()
demo = gr.Interface(fn=translate,
                    inputs=[gr.Textbox(label="English Text:", lines=2)],
                    outputs=[gr.Audio(label="Rendered Audio:")],

                    allow_flagging="never",
                    #Here we introduce a new tag, examples, easy to use examples for your application
)
demo.launch(share=True)
```

```python
# Close all Gradio interfaces after launching the demo
gr.close_all()
```
